## Process the manually selected images

```python
from pathlib import Path

import geopandas as gpd
import pandas as pd
```

```python
outlines_fp = '../data/outlines/paul_et_al_2020/c3s_gi_rgi11_s2_2015_v2.shp'
gl_df = gpd.read_file(outlines_fp)
gl_df
```

```python
# the next csv file contains the cloud coverage percentage for all the images from the inventory years 
# (obtained by running with C = S2_ALPS in config.py, assuming all the images were downloaded beforehand)

df_clouds = pd.read_csv('../data/external/wd/s2_alps/inv/aux_data/cloud_stats.csv')
df_clouds['date'] = df_clouds.fp_img.apply(lambda fp: pd.to_datetime(Path(fp).name[:8]).strftime('%Y-%m-%d'))
display(df_clouds.drop(columns='fp_img'))
df_clouds['k'] = df_clouds.entry_id + '_' + df_clouds.date

sdf_clouds = df_clouds[df_clouds.cloud_p_gl_b50m_v1 <= 0.3]
print(len(sdf_clouds))
```

```python
# build a dataframe with all the initial glaciers and add the cloud coverage percentage
df_dates_initial = gl_df[gl_df.area_km2 >= 0.1][['entry_id', 'date_inv']].copy()
df_dates_initial.date_inv = df_dates_initial.date_inv.apply(lambda x: x.strftime(f"%Y-%m-%d"))
df_dates_initial = df_dates_initial.merge(df_clouds[['entry_id', 'date', 'cloud_p_gl_b50m_v1']],
                                          left_on=('entry_id', 'date_inv'), right_on=('entry_id', 'date'), how='left')
assert not df_dates_initial.cloud_p_gl_b50m_v1.isna().any()
display(df_dates_initial)
```

```python
df_to_remove = pd.read_csv('../data/inv_images_qc/gl_seg - imgs_to_remove.csv')
df_to_remove['comment'] = df_to_remove.apply(lambda r: ' & '.join([c for c in df_to_remove.columns if r[c] == 'x']),
                                             axis=1)
df_to_remove = df_to_remove[['entry_id', 'date', 'comment']]
print(f"len(df_to_remove) = {len(df_to_remove)}")
# make sure the dates are within the inventory
assert len(df_to_remove.merge(df_dates_initial, on=('entry_id', 'date'), how='inner')) == len(df_to_remove)
display(df_to_remove)

df_to_replace = pd.read_csv('../data/inv_images_qc/gl_seg - imgs_to_replace.csv')
df_to_replace['comment'] = df_to_replace.apply(lambda r: ' & '.join([c for c in df_to_replace.columns if r[c] == 'x']),
                                               axis=1)
df_to_replace = df_to_replace[['entry_id', 'date', 'comment']]
# make sure the dates are within the sdf_clouds (i.e. the cloud coverage is below 30%)
assert len(df_to_replace.merge(sdf_clouds, on=('entry_id', 'date'), how='inner')) == len(df_to_replace)
print(f"len(df_to_replace) = {len(df_to_replace)}")
display(df_to_replace)

assert len(set(df_to_remove.entry_id) & set(df_to_replace.entry_id)) == 0
```

```python
# check how many glaciers we would drop if we don't do any change
glaciers_with_too_many_clouds = set(df_dates_initial.entry_id[df_dates_initial.cloud_p_gl_b50m_v1 > 0.3])
glaciers_to_manually_remove = set(df_to_remove.entry_id)
print(f"#glaciers with more than 30% cloud coverage = {len(glaciers_with_too_many_clouds)}")
print(f"#glaciers to manually remove = {len(glaciers_to_manually_remove)}")
glaciers_to_manually_remove -= glaciers_with_too_many_clouds
print(
    f"#glaciers to manually remove (after excluding those with more than 30% cloud coverage) = {len(glaciers_to_manually_remove)}")
glaciers_to_remove = glaciers_with_too_many_clouds | glaciers_to_manually_remove
print(f"#glaciers to be removed in total = {len(glaciers_to_remove)}")
```

```python
# build a dataframe and replace the manually selected dates; then add again the cloud coverage percentage
df_dates_final = df_dates_initial[['entry_id', 'date_inv']].copy()
df_dates_final['date'] = df_dates_final.date_inv
df_dates_final.set_index('entry_id', inplace=True)
df_dates_final.loc[df_to_replace.entry_id, 'date'] = df_to_replace.date.values
df_dates_final = df_dates_final.merge(df_clouds[['entry_id', 'date', 'cloud_p_gl_b50m_v1']], on=('entry_id', 'date'),
                                      how='left')
assert not df_dates_final.cloud_p_gl_b50m_v1.isna().any()
display(df_dates_final)

glaciers_with_too_many_clouds_v2 = set(df_dates_final.entry_id[df_dates_final.cloud_p_gl_b50m_v1 > 0.3])
glaciers_to_manually_remove_v2 = set(df_to_remove.entry_id) - set(glaciers_with_too_many_clouds_v2)
print(f"#glaciers with more than 30% cloud coverage = {len(glaciers_with_too_many_clouds_v2)}")
print(
    f"#glaciers to manually remove (after excluding those with more than 30% cloud coverage) = {len(glaciers_to_manually_remove_v2)}")
glaciers_to_remove_v2 = glaciers_with_too_many_clouds_v2 | glaciers_to_manually_remove_v2
print(f"#glaciers to be removed in total = {len(glaciers_to_remove_v2)}")
print(f"#glaciers that were 'saved' by replacing the dates: {len(glaciers_to_remove - glaciers_to_remove_v2)}")

# remove the selected glaciers
df_dates_final = df_dates_final[~df_dates_final.entry_id.isin(list(glaciers_to_remove_v2))]

# check how many of the remaining glaciers have a changed date
print(f"#glaciers with a different date in the end = {(df_dates_final.date != df_dates_final.date_inv).sum()}")

# check how many dates were optionally changed (i.e. exclude the problematic glaciers)
print(
    f"#glaciers with a different date in the end (excluding the removed ones) = {len(df_dates_final[~df_dates_final.entry_id.isin(glaciers_to_remove) & (df_dates_final.date != df_dates_final.date_inv)])}")

df_dates_final.to_csv('../data/inv_images_qc/final_dates.csv', index=False)
df_dates_final
```
